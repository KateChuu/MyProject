```python
import pandas as pd
df = pd.read_csv('resources/week-2/datasets/presidents.csv')
df.head()
```

```python
# let's start with cleaning up that name into first name and last name, we're going to tackle this with a regex. 
# we want to create two new columns and apply a regex to the projection of the president column.

# creating a new column named 'First'
df["First"] = df["President"]
# the pattern is a whitespave followed by any number of characters, and we replace it with null
df["First"] = df["First"].replace("[ ].*", "", regex=True)
df.head()
```

But it's slow since we had to make a full copy of the column and then go through and update strings. There are a few other ways . The most general one first and that's the apply function. 

```python
# So we use del to drop the column 'First' immediately in place. No copy of a data frame being returned.
del(df['First'])
df.head()
```

```python
# the apply function on a data frame will take some arbitrary function you have written
# and apply it to either a series, a single column or a data frame across all rows or columns.
# Let's write a function which just splits a string into two pieces using a single row of data.

# row is a single series objectwhich is a single row indexed by column values
def splitname(row):
    # extracting the firstname and create a new entry in the series
    row['First'] = row['President'].split(" ")[0]
    # do the same with the last word in the string, -1 means counting from the last word
    row['Last'] = row['President'].split(" ")[-1]
    return row

df = df.apply(splitname, axis='columns') # axis='columns' means applying function to each row.
df.head()
```

```python
# dropping the first name and the last name
del(df['First'])
del(df['Last']) 
df.head()
```

Another handy function called extract takes a regular expression as input and specifically requires we to set capture groups that correspond to the output columns that we're interested in.

```python
# a regular expression that returned groups and had just the first name and last name in it
# we'll match 3 groups but only return 2, the first and the last name, it won't capture the characters after ?:
pattern = "(^[\w]*)(?:.* )([\w]*$)"

# the extract function is built into the str attribute of the series object
df['President'].str.extract(pattern).head()
```

```python
# it looks pretty but the column names, we can name the column with pattern 
pattern = "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"
names = df['President'].str.extract(pattern).head()
names
```

```python
# then we can copy these into our main dataframe
df['First'] = names['First']
df['Last'] = names['Last']
df.head()
```

Now let's move on to clean up that Born column, we'll first get rid of anything that isn't in the pattern of month day and year.

```python
df['Born'] = df['Born'].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df['Born'].head()
```

```python
# The type of this column is actually an object, but pandas actually has really interesting date-time features.
# If we were building this out further, we would actually update this column to write the data type as well.
df['Born'] = pd.to_datetime(df['Born'])
df['Born'].head()
```
