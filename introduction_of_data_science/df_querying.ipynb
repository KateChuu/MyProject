Boolean masking

A Boolean mask is an array which can be thought of as a one dimension like a series, or two-dimensions like a DataFrame, where each of the values of the array are either true or false. This array is essentially overlaid on top of the other data structure that we're querying, and any cell aligned with the true value will be admitted into our final result, and any cell aligned with the false value will not.

```python
import pandas as pd
# set column 0 to be the serial numbers
df = pd.read_csv('resources/week-2/datasets/Admission_Predict.csv', index_col=0)

# turning it into lower case and stripping out whitespace
df.columns = [x.lower().strip() for x in df.columns]
df.head()
```

```python
# we are interested in those students that have a chance higher than 0.7

# the steps:
# 1. projecting the chance of admit column using the index operator
# 2. applying the > 0.7 operator
# 3. the resulte returned in a boolean series, where the value of each cell is either True or False

admit_mask = df['chance of admit'] > 0.7
admit_mask
```

```python
# filtering the False values we don't want by using the .where() function
df.where(admit_mask).head()
```

```python
# the resulting DataFrame keeps the original index values, and only data which met the condition was retained. 
# the rows which did not meet the condition have NaN data instead, but these rows weren't dropped from our dataset. 

# we use the dropna() function to drop the NaN data
df.where(admit_mask).dropna().head()
```

```python
# a shorthand syntax which combines where() and dropna(), the result is same as the anove one
df[df['chance of admit'] > 0.7].head()
```

```python
# This indexing operator does two things

# 1. it can be called with a string parameter to project a single column
df['gre score'].head()
```

```python
# or sending it a list of columns as string
df[['gre score','toefl score']].head()
```

```python
# 2. sending it a boolean mask
df[df['gre score']>320].head()
```

each of these is mimicing functionality from either .loc() or .where().dropna()

Combining multiple Boolean masks

```python
# If both masks must be True, we use 'and', BUT it does NOT work!!!!
(df['chance of admit'] > 0.7) and (df['chance of admit'] < 0.9)
```

The problem is that we have series objects, and python doesn't know how to compare two series using 'and' or 'or'. Instead, the pandas authors have overwritten the pipe | and ampersand & operators to handle this 

```python
(df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)
```

```python
# a common error is to do boolean comparisons using the & operator but not putting parentheses () around
df['chance of admit'] > 0.7 & df['chance of admit'] < 0.9
```

```python
# another way is to use the built-in functions, so that we don't have to put parentheses around
df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)
```

```python
# we can also chain them
df['chance of admit'].gt(0.7).lt(0.9)
```
