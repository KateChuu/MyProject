# Naive Bayes classification
- based on the Bayes theorem
- it assumes that data follows the normal (Gaussian) distribution.
- means that the data points within each class are assumed to have a mean and standard deviation, and the algorithm uses these statistical properties to make predications about new data points.

### What is the meaning of 'naive'?
- comes from the suumption that all features of the data are independent of each other.
- so the algorithm doesn't consider the relationship between different features when making predictions.
- only consider the individual probabilities of each feature for each class.

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=2, cluster_std=1.5)
# n_features=2: each data has 2 features
# centers=2: we want 2 clusters
```

```python
# visualize the fitting of Gaussian Naive Bayes model to the generated data

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=100, n_features=2, random_state=2, cluster_std=1.5)

fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
ax.set_title('Naive Bayes Model', size=14)

xlim = (-8, 8)
ylim = (-15, 5)

xg = np.linspace(xlim[0], xlim[1], 60)
yg = np.linspace(ylim[0], ylim[1], 60) 

xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T

for label, color in enumerate(['red', 'blue']):
    mask = (y==label)
    mu, std = X[mask].mean(0), X[mask].std(0)
    P = np.exp(-0.5 * (Xgrid - mu) ** 2/std ** 2).prod(1) # probability of being in which group?????
    Pm = np.ma.masked_array(P, P<0.003)
    ax.pcolorfast(xg, yg, Pm.reshape(xx.shape), alpha=0.5, cmap=color.title() + 's')
    ax.contour(xx, yy, P.reshape(xx.shape), levels=[0.01, 0.1, 0.5, 0.9], colors=color, alpha=0.2)

ax.set(xlim=xlim, ylim=ylim)
```

```python
# generate new data for predicting the label using Gaussian Naive Bayes model

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y)
```

```python
rng = np.random.RandomState(42)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2) # data size = 2000*2
ynew = model.predict(Xnew)
```
